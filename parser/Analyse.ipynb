```python
import os
import pandas as pd
import re
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
def get_procent_tactic(string):
    try:
        result = re.search('\d{1,3}', string).group()
        return result
    except:
        return -1
```

```python
df = pd.read_csv('../data/concat_dataset', index_col=0)
df.drop_duplicates(inplace=True)
```

```python
#creates datasets with bots and without bots
df_with_bots = df[(df.player_1 == 'Спарринг-Бот') | (df.player_2 == 'Спарринг-Бот')]
df_players = df_with_bots = df[(df.player_1 != 'Спарринг-Бот') & (df.player_2 != 'Спарринг-Бот')]
```

```python
# create new column with procent tactic
df_players['tactic_1p'] = map(lambda x: get_procent_tactic(x), df_players['tactic_1'])
df_players['tactic_2p'] = map(lambda x: get_procent_tactic(x), df_players['tactic_2'])

df_with_bots['tactic_1p'] = map(lambda x: get_procent_tactic(x), df_with_bots['tactic_1'])
df_with_bots['tactic_2p'] = map(lambda x: get_procent_tactic(x), df_with_bots['tactic_2'])
```

```python
# take columns for model machine learning
df_players.drop(labels=['time', 'viewers', 'id_match', 'tactic_1', 'tactic_2', 'player_1', 'player_2'], axis=1, inplace=True)

```

```python
# update indexes
df_players.index = range(len(df_players))

df_with_bots.index = range(len(df_players))
```

```python
# delete rows with missmatch opponent
df_players = df_players[(df_players['tactic_1p'] != -1) & (df_players['tactic_2p'] != -1)]

# delete rows with missmatch opponent
df_with_bots = df_players[(df_players['tactic_1p'] != -1) & (df_players['tactic_2p'] != -1)]
```
